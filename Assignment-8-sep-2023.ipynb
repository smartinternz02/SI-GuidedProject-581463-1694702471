```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

Load Dataset

```python
df = sns.load_dataset('car_crashes')
```

```python
df.head()
```

Scatterplot

```python
sns.scatterplot(x="total",y="alcohol",data=df)
```

Inference: from the plot we can say that as the percentage of drivers that were alcohol-impaired increases the total number of drivers involved in fatal collisions also increases.

```python
sns.lineplot(x="speeding",y="total",data=df,errorbar=None)
```

Inference: from the plot we can say that total number of drivers involved in fatal collisions may rise or fall as speeding increases  

```python
df1 = df.head(10)
sns.relplot(x="total",y="alcohol",data=df1,hue="abbrev")
```

Inference: In state AR maximum drivers were alcohol-impaired and were involved in fatal collisions.(for first 10 records)  

```python
df2 = df.head(20)
sns.barplot(data=df2,x="abbrev",y="ins_premium",ci=None)
```

Inference: LA has the most car insurance premiums among the other states.(for first 20 records)

```python
sns.countplot(x="abbrev",data=df1)
```

Inference: each state appears only once in the dataset.

```python
df = df.drop(columns=["abbrev"])
corr = df.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

Inference: total and no_previous are the most correlated columns.

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])  #[left,bottom,width,height]
labels = df1["abbrev"]
axes1.pie(df1["ins_losses"],labels=labels,autopct="%0.2f%%",colors=["orange","pink","green","blue"])
axes1.legend()
```

Inference: In CT, losses incurred by insurance companies for collisions per insured driver(ins_losses) distribution is highest . (for first 10 records)
